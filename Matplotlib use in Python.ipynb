```python
days=[1,2,3,4,5,6,7]
max_t = [31,45,49,32,44,36,49]
min_t = [19,20,21,22,23,24,25]
avg_t = [45,45,45,45,45,45,45]
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label = 'min')
plt.plot(days,avg_t,label = 'avg')
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('week temperature')
plt.legend(loc ='lower right')
```

```python
#bar graph

company =['google','amazon','msft','fb']
revenue= [90,136,89,27]
profit =[40,20,14,21]
plt.bar(company,revenue)
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('us tech stocks')
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
plt.hist(df["Age"])
```

```python
#pie-charts
exp=[1400,600,300,410,250]
exp_labels=['home','food','phone','car','other']
plt.pie(x=exp,labels=exp_labels)
```

```python
plt.pie(exp , labels=exp_labels , radius=1,autopct='%0.1f%%')
plt.plot()
```

```python
exp=[1400,600,300,410,250]
exp_labels=['home','food','phone','car','other']
plt.pie(exp, labels = exp_labels ,radius=1,autopct='%0.2f%%',explode=[0.1,0,0,0.2,0])
plt.plot() 
```

```python
embarked_group=df.groupby('Embarked').agg(Fare_collected=('Fare','sum'))
print(embarked_group.index.values)
print(embarked_group['Fare_collected'].values)

plt.pie(labels = embarked_group.index.values,
        x = embarked_group['Fare_collected'].values,
       
       shadow = True,
       explode = [0.2,0,0],
       counterclock = False,
       colors= ['b','y','g'],
       autopct = '%1.2f%%')   


plt.title('________ Pie Chart ________')
plt.show()
```

```python
year = [2014,2015,2016,2017]
income =[40000,45000,50000,46000]
expence = [2800,300,2800,3400]

plt.subplots_adjust(hspace=0.4,wspace=0.5)  #it seperates the graphs/provides space between the graphs

plt.subplot(1,2,1)
plt.bar(year,income)
plt.xlabel('xyz')

plt.subplot(1,2,2)
plt.bar(year,expence, color='orange')
plt.xlabel('abc')
```
